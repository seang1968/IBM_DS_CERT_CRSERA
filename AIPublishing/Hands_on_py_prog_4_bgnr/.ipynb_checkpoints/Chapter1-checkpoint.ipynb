## Lists

```python
colors =["Red","Green","Blue","Orange"] 
print(colors) 
print(type(colors))
```

```python
mixed_list=[20,"Male","USA",True] 
print(mixed_list)
```

#### Acessing list items

```python
colors =["Red","Green","Blue","Orange"] 
print(colors[1])
```

```python
for col in colors: 
    print(col)
```

#### Adding Elements to a List

```python
colors =["Red","Green","Blue","Orange"] 
colors.append("Black") 
print(colors)
```

#### Updating a List

```python
colors =["Red","Green","Blue","Orange"] 
colors[1]="Yellow" 
print(colors)
```

```python
#This errs out.. You can not add a new item this way... It works with Dictionaries
colors[4]="Yellow" 
print(colors)
```

#### Deleting a List Item

```python
colors =["Red","Green","Blue","Orange"] 
colors.remove("Green") 
print(colors)
```

```python
#Pop removes item at index
colors =["Red","Green","Blue","Orange"] 
colors.pop(2) 
print(colors)
```

## Dictionaries

```python
country_capitals={"England":"London", "France":"Paris", "Japan":"Tokyo", "China":"Beijing"} 
print(country_capitals) 
print(type(country_capitals))
```

#### Accessing Dictionary Items

```python
# can not access with index.. the index is the key itself
country_capitals[0]
```

```python
print(country_capitals["Japan"])
```

```python
for k,v in country_capitals.items(): 
    print(k+" ->"+v)
```

```python
for k in country_capitals.keys(): 
    print(k)
```

```python
for k in country_capitals.items(): 
    print(k)
```
